```python
# Cargar las hojas de Excel en DataFrames
hoja0 = pd.read_excel('/content/drive/MyDrive/PROYECTO_TALENTO_TECH/Base ventas historicas tanques.xlsx', sheet_name='2019')
hoja1 = pd.read_excel('/content/drive/MyDrive/PROYECTO_TALENTO_TECH/Base ventas historicas tanques.xlsx', sheet_name='2020')
hoja2 = pd.read_excel('/content/drive/MyDrive/PROYECTO_TALENTO_TECH/Base ventas historicas tanques.xlsx', sheet_name='2021')
hoja3 = pd.read_excel('/content/drive/MyDrive/PROYECTO_TALENTO_TECH/Base ventas historicas tanques.xlsx', sheet_name='2022')
hoja4 = pd.read_excel('/content/drive/MyDrive/PROYECTO_TALENTO_TECH/Base ventas historicas tanques.xlsx', sheet_name='2023')
hoja5 = pd.read_excel('/content/drive/MyDrive/PROYECTO_TALENTO_TECH/Base ventas historicas tanques.xlsx', sheet_name='2024')
```

```python
# Seleccionar las columnas deseadas de cada hoja
columnas_deseadas = ['Fecha', 'CODIGO SAP', 'NOMBRE ITEM SAP', 'BODEGA MOVIMIENTO','VENTAS NETAS UNIDADES']

# Combinar las tres hojas en una sola conservando solo las columnas deseadas
nueva_hoja = pd.concat([hoja0[columnas_deseadas], hoja1[columnas_deseadas], hoja2[columnas_deseadas], hoja3[columnas_deseadas], hoja4[columnas_deseadas], hoja5[columnas_deseadas]], ignore_index=True)

# Ordenar el DataFrame resultante por la columna 'Fecha'
nueva_hoja = nueva_hoja.sort_values(by='Fecha')

#Muestreame el nuevo data frame
nueva_hoja.head()
```

#Preprocesamiento_1

```python
nueva_hoja['VENTAS NETAS UNIDADES'] = nueva_hoja['VENTAS NETAS UNIDADES'].astype(float)
```

```python
# Eliminar filas con ventas netas igual a cero

nueva_hoja = nueva_hoja[(nueva_hoja['VENTAS NETAS UNIDADES'] > 0)]
```

```python
nueva_hoja = nueva_hoja.sort_values(by='VENTAS NETAS UNIDADES', ascending=True)
#Mostrar dataframe
nueva_hoja.head()
```

```python
#Cargar librerias
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Suponiendo que ya tienes el DataFrame 'nueva_hoja' con las columnas necesarias

# Reorganizar los datos utilizando pivot_table para agrupar por fecha y código SAP
pivot_data_codigos = nueva_hoja.pivot_table(index='Fecha', columns='CODIGO SAP', values='VENTAS NETAS UNIDADES', aggfunc='sum')

# Reemplazar NaN por 0 (si se desea)
pivot_data_codigos = pivot_data_codigos.fillna(0)

# Restablecer el índice para que 'Fecha' sea una columna y los códigos sean las demás columnas
pivot_data_codigos = pivot_data_codigos.reset_index()

# Añadir columnas para el mes, semana, número de día de la semana y nombre del día
pivot_data_codigos['Mes'] = pivot_data_codigos['Fecha'].dt.month
pivot_data_codigos['Semana'] = pivot_data_codigos['Fecha'].dt.isocalendar().week
pivot_data_codigos['DiaSemanaNum'] = pivot_data_codigos['Fecha'].dt.weekday  # 0 = Lunes, 6 = Domingo
pivot_data_codigos['DiaSemanaNombre'] = pivot_data_codigos['Fecha'].dt.day_name()  # Lunes, Martes, ...

# Mostrar el resultado
pivot_data_codigos.head()
```

```python
# Reorganizar los datos utilizando pivot_table para agrupar por fecha y nombre del item SAP
pivot_data_nombre = nueva_hoja.pivot_table(index='Fecha', columns='NOMBRE ITEM SAP', values='VENTAS NETAS UNIDADES', aggfunc='sum')

# Reemplazar NaN por 0 (si se desea)
pivot_data_nombre = pivot_data_nombre.fillna(0)

# Restablecer el índice para que 'Fecha' sea una columna y los códigos sean las demás columnas
pivot_data_nombre = pivot_data_nombre.reset_index()

# Añadir columnas para el mes, semana, número de día de la semana y nombre del día
pivot_data_nombre['Mes'] = pivot_data_nombre['Fecha'].dt.month
pivot_data_nombre['Semana'] = pivot_data_nombre['Fecha'].dt.isocalendar().week
pivot_data_nombre['DiaSemanaNum'] = pivot_data_nombre['Fecha'].dt.weekday  # 0 = Lunes, 6 = Domingo
pivot_data_nombre['DiaSemanaNombre'] = pivot_data_nombre['Fecha'].dt.day_name()  # Lunes, Martes, ...

# Mostrar el resultado
pivot_data_nombre.head()
```

Normalizando para referencias de productos de 2000 L

1. Normalizar TANQ por TANQUE
2. Cambiar LTS por L
3. 2000L por 2000 LTS
4. Eliminar espacios extra
5. JUEG0 por JUEGO


*Falta estandarizar los kits*


```python
normalizacion = nueva_hoja
```

```python
# Paso 1: Realizar las normalizaciones en `NOMBRE ITEM SAP`
normalizacion['NOMBRE ITEM SAP'] = (
    normalizacion['NOMBRE ITEM SAP']
    .str.replace(r'\bTANQ\b', 'TANQUE', regex=True)  # Cambiar "TANQ" por "TANQUE"
    .str.replace(r'\bLTS\b', 'L', regex=True)  # Cambiar "LTS" por "L"
    .str.replace(r'(\d+)L\b', r'\1 L', regex=True)  # Asegurar formato "<CAPACIDAD> L"
    .str.replace(r'\bJUEG0\b', 'JUEGO', regex=True)  # Cambiar "JUEG0" (con cero) por "JUEGO"
    .str.replace(r'JUEGOTANQUE', 'JUEGO TANQUE', regex=True)  # Asegurar espacio entre "JUEGO" y "TANQUE"
    .str.replace(r'\s+', ' ', regex=True)  # Remover espacios extra
    .str.strip()  # Quitar espacios al inicio y al final
)

# Paso 2: Calcular las ventas totales para cada elemento único en 'NOMBRE ITEM SAP'
ventas_totales = (
    normalizacion.groupby('NOMBRE ITEM SAP')['VENTAS NETAS UNIDADES']
    .sum()
    .reset_index()
)

# Paso 3: Guardar el DataFrame en un archivo Excel
ventas_totales.to_excel('ventas_totales_por_producto.xlsx', index=False)

print("El archivo 'ventas_totales_por_producto.xlsx' se ha generado exitosamente con los productos normalizados.")
```
